```python
import requests
import pandas as pd
import time

# Вставьте ваш ключ API
API_KEY = ''

# URL для доступа к API NASA Astronomy Picture of the Day
BASE_URL = 'https://api.nasa.gov/planetary/apod'
```

#### Проверяем подключение

```python
# Параметры запроса
params = {
    'api_key': API_KEY,
    'count': 1  # Получаем одно изображение на запрос
}

# Запрос к API
response = requests.get(BASE_URL, params=params)

# Проверим статус запроса и выведем результат
if response.status_code == 200:
    print("Запрос успешен!")
    print(response.json())  # Выводим данные, которые мы получили
else:
    print(f"Ошибка при получении данных: {response.status_code}")
```

### Собираем датасет

```python
# Параметры запроса
params = {
    'api_key': API_KEY,
    'count': 100  # Получаем 100 изображений за один запрос
}

# Список для хранения данных
all_data = []

# Количество запросов (например, 50 запросов, чтобы получить 5000 строк)
num_requests = 50

for i in range(num_requests):
    # Запрос к API
    response = requests.get(BASE_URL, params=params)
    
    # Проверка на успешность запроса
    if response.status_code == 200:
        print(f"Запрос {i+1} успешен!")
        data = response.json()  # Получаем данные
        
        # Добавляем данные в список
        all_data.extend(data)
    else:
        print(f"Ошибка при получении данных: {response.status_code}")
    
    # Задержка между запросами
    time.sleep(1)

# После завершения цикла получаем 5000+ строк данных
print(f"Получено {len(all_data)} записей.")

# Преобразуем данные в DataFrame для дальнейшего анализа
import pandas as pd

df_nasa = pd.DataFrame(all_data)

# Посмотрим на первые строки
df_nasa.head()
```

#### Проведём EDA-анализ

```python
# Проверим информацию о столбцах и типах данных
print(df_nasa.info())

# Посчитаем пропущенные значения
print(df_nasa.isnull().sum())
```

```python
# Преобразуем date в формат datetime
df_nasa['date'] = pd.to_datetime(df_nasa['date'])

# Посмотрим уникальные значения media_type
print(df_nasa['media_type'].value_counts())

# Заполним пропуски в url значением из hdurl (если есть)
df_nasa['url'].fillna(df_nasa['hdurl'], inplace=True)

# Заполним copyright строкой "Unknown", если пусто
df_nasa['copyright'].fillna('Unknown', inplace=True)

# Проверим, остались ли пропуски
print(df_nasa.isnull().sum())
```

После подготовки и очистки (где возможно) данных приступаем к выполнению задач.

### Задача 1

**Исследование динамики публикаций**   
   - Определить, в какие месяцы чаще всего публикуются материалы (есть ли в датасете сезонные/временные искажения?)  
   - Выявить закономерности в частоте публикаций  


```python
df_nasa['month'] = df_nasa['date'].dt.month # Делаем столбец с месяцем
import matplotlib.pyplot as plt

# Считаем количество записей по месяцам
month_counts = df_nasa['month'].value_counts().sort_index()

# Построим график
month_counts.plot(kind='bar', figsize=(10,6), color='skyblue')
plt.title("Распределение записей по месяцам")
plt.xlabel("Месяц")
plt.ylabel("Количество записей")
plt.xticks(range(12), ['Январь', 'Февраль', 'Март', 'Апрель', 'Май', 'Июнь', 'Июль', 'Август', 'Сентябрь', 'Октябрь', 'Ноябрь', 'Декабрь'], rotation=45)
plt.show()
```

```python
month_summary = df_nasa.groupby('month').size().reset_index(name='count')
month_summary.to_csv('month_summary.csv', index=False)  # Сохраняем в CSV
```

```python
# Среднее количество записей по месяцам
print(f"Среднее количество записей по месяцам: {month_counts.mean()}")

# Редкость типов медиа по месяцам
media_by_month = df_nasa.groupby(['month', 'media_type']).size().unstack(fill_value=0)
media_by_month.to_csv('media_by_month.csv', index=True)
```

**Вывод**: Количество записей стабильно распределено по месяцам, с небольшими колебаниями. Данные довольно равномерно распределены по году, в среднем 416.6666666666667 записей в месяц. Это говорит о том, что в датасете отсутствуют сезонные и временные искажения, сбор данных равномерен.

```python
# Диаграмма для типов медиа по месяцам
media_by_month.plot(kind='bar', stacked=False, figsize=(10, 6))
plt.title('Типы медиа по месяцам')
plt.xlabel('Месяц')
plt.ylabel('Количество')
plt.xticks(rotation=0)
plt.legend(title='Тип медиа')
plt.show()
```

**Вывод**: Почти все записи относятся к изображениям, видео на втором месте. Другие типы медиа составляют минимальную долю данных.

### Задача 2

 **Анализ типов контента**  
   - Определить, какой тип контента преобладает (изображения или видео)  
   - Проследить, как изменялось соотношение типов контента по годам  

```python
import matplotlib.pyplot as plt

# Срез для распределения типов контента
media_type_counts = df_nasa['media_type'].value_counts()

# Построение круговой диаграммы
plt.figure(figsize=(7, 7))
plt.pie(media_type_counts, labels=media_type_counts.index, autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99'])
plt.title('Распределение типов контента')
plt.axis('equal')  # Сделать круговую диаграмму круглой
plt.show()
```

**Вывод**: Как видно из диаграммы, абсолютное большинство файлов – изображения (96.6%). Видео на втором месте примерно с 3.1%. Возможно, следующий график поможет понять, почему разница настолько большая.

```python
# Срез для изменений типов контента по годам

# Создадим новый столбец 'year', извлекая год из столбца 'date'
df_nasa['year'] = df_nasa['date'].dt.year

# Проверим результаты
print(df_nasa[['date', 'year']].head())

media_by_year = df_nasa.groupby(['year', 'media_type']).size().unstack().fillna(0)
print("Изменения типов контента по годам:")
print(media_by_year)
media_by_year.to_csv('media_by_year.csv', index=True)
```

```python
# Построение линейного графика изменения типов контента по годам
media_by_year.plot(kind='line', figsize=(10, 6), marker='o', color=['skyblue', 'lightgreen', 'salmon'])
plt.title('Изменение типов контента по годам')
plt.xlabel('Год')
plt.ylabel('Количество')
plt.legend(title='Тип контента')
plt.xticks(rotation=45)
plt.show()
```

**Вывод**: График показывает тренд использования тех или иных типов медиа. Формат видео впервые был использован в 2010, в то время как фото активно использовались с 1995. Можно предположить, что такое преобладание фото- над видео-форматом связано с технологическими сложностями. Впервые технология записи видео, судя по всему, была использована в 2010.

### Задача 3

 **Анализ авторов публикаций**  
   - Выявить самых активных авторов изображений  
   - Определить долю изображений, принадлежащих NASA  


```python
images_df = df_nasa[df_nasa['media_type'] == 'image']

# Фильтруем данные, исключая "Unknown" авторов
images_df_without_unknown = images_df[images_df['copyright'] != 'Unknown']

# Группируем по автору и считаем количество публикаций
active_authors = images_df_without_unknown['copyright'].value_counts()

# Отобразим топ 10 самых активных авторов
top_active_authors = active_authors.head(10)
print("Топ 10 самых активных авторов изображений:")
print(top_active_authors)
```

```python
# Фильтруем только изображения, автором которых является NASA
nasa_images = images_df[images_df['copyright'].str.contains('NASA', case=False, na=False)]

# Доля изображений, принадлежащих NASA
nasa_percentage = (len(nasa_images) / len(images_df)) * 100
print(f"Доля изображений, принадлежащих NASA: {nasa_percentage:.2f}%")
```

```python
# Сохраняем топ авторов в CSV
top_active_authors.to_csv('top_active_authors.csv', header=True)

# Сохраняем изображения, принадлежащие NASA, в CSV
nasa_images.to_csv('nasa_images.csv', index=False)
```

```python
# График для самых активных авторов
plt.figure(figsize=(10, 6))
top_active_authors.plot(kind='bar', color='skyblue')
plt.title('Топ 10 самых активных авторов изображений')
plt.ylabel('Количество публикаций')
plt.xlabel('Автор')
plt.xticks(rotation=45)
plt.show()

# График для доли изображений NASA
labels = ['NASA', 'Другие']
sizes = [len(nasa_images), len(images_df) - len(nasa_images)]
colors = ['#ff9999', '#66b3ff']
plt.figure(figsize=(7, 7))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Доля изображений, принадлежащих NASA')
plt.axis('equal')
plt.show()
```

**Вывод**: Список самых популярных авторов можно увидеть на графике сверху. 
На графике снизу: Меньше 1% изображений под лицензией NASA.

### Задача 4

**Анализ тематики публикаций**  
   - Определить, какие астрономические объекты (планеты, галактики, звезды) упоминаются чаще всего  
   - Исследовать, как менялась популярность различных объектов во времени
   - Проанализировать, как медиа отражают природные явления на примере упоминания комет по годам.  


```python
# Список астрономических объектов, которые мы будем искать в пояснениях
astronomical_objects = ['planet', 'galaxy', 'star', 'nebula', 'asteroid', 'comet', 'supernova']

# Считаем упоминания каждого объекта по годам
object_mentions_by_year = pd.DataFrame()

for obj in astronomical_objects:
    object_mentions_by_year[obj] = df_nasa[df_nasa['explanation'].str.contains(obj, case=False, na=False)].groupby('year').size()

# Сохраняем в CSV
object_mentions_by_year.to_csv('object_mentions_by_year.csv', index=True)

# Построим линейный график для изменений по годам
object_mentions_by_year.plot(kind='line', figsize=(12, 8))
plt.title('Популярность астрономических объектов по годам')
plt.xlabel('Год')
plt.ylabel('Количество упоминаний')
plt.legend(title='Объекты')
plt.show()

# Выводим данные
display(df_nasa)
```

**Вывод**: Моя гипотеза заключалась в том, что график мог бы позволить отследить природные явления по частоте снимков с тематическим словом в определённый год по пикам. При поиске природных явлений по ключевым словам и году в Интернете я не нашла точные совпадения. Гипотеза не подтвердилась.

```python
# Гистограмма упоминаний слова "комета" по годам
comet_mentions = df_nasa[df_nasa['explanation'].str.contains('comet', case=False, na=False)]

# Извлекаем год из столбца даты
comet_mentions['year'] = comet_mentions['date'].dt.year

# Построим гистограмму по годам
plt.figure(figsize=(10, 6))
plt.hist(comet_mentions['year'], bins=range(comet_mentions['year'].min(), comet_mentions['year'].max() + 1), edgecolor='black')
plt.title('Количество упоминаний комет по годам')
plt.xlabel('Год')
plt.ylabel('Количество упоминаний')
plt.xticks(range(comet_mentions['year'].min(), comet_mentions['year'].max() + 1), rotation=90)  # Вертикальные метки
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()  # Это поможет, чтобы метки не выходили за пределы графика
plt.show()

```

**Вывод:** Некоторые всплески действительно происходят в одно время с большими кометами: Хякутакэ (1996), комета Хейла-Боппа (1997), Комета Макнота (2007). Но есть и всплески, которые не связаны с большими кометами (например, 2013-2014). Гипотеза подтвердилась частично.
